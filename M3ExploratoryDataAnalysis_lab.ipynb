## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(df['ConvertedComp'])
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
m = df['ConvertedComp'].median()
df['ConvertedComp'].fillna(m,inplace=True)
Q1 = np.percentile(df['ConvertedComp'],25)
Q3 = np.percentile(df['ConvertedComp'],75)
IQR = Q3 - Q1
print(f'Q1 = {Q1} , Q3 = {Q3} , IQR = {IQR}')
```

Find out the upper and lower bounds.


```python
# your code goes here
Lower_bound = Q1 - (1.5 * IQR)
Upper_bound = Q3 + (1.5 * IQR)
print(f'Lower bound = {Lower_bound} , Upper bound = {Upper_bound}')
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
df[df['ConvertedComp']>Upper_bound]['ConvertedComp'].count()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
new_df = df.copy()

new_df['ConvertedComp'] = np.where(df['ConvertedComp']>Upper_bound,Upper_bound,new_df['ConvertedComp'])
```

```python
sns.boxplot(new_df['ConvertedComp'])
plt.show()
```
